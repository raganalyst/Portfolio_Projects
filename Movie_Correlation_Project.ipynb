```python
# Import Librarirs

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Read in the data
df = pd.read_csv(r"C:\Users\91620\OneDrive\Desktop\Project\moviesData_for Python.csv")
```

```python
# look at Data
df
```

```python
# See if we have any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Converting all nulls in the Gross and Budget column into Zero

df['budget'] = df['budget'].fillna(0)

df['gross'] = df['gross'].fillna(0)

df['rating'] = df['rating'].fillna(0)

df['votes'] = df['votes'].fillna(0)
# To check

print(df[['budget','gross','rating']].isnull().sum())
```

```python
##df = df.dropna()
```

```python
# Data Types of columns

print(df.dtypes)
```

```python
# Change data type of columns

df['gross'] = df['gross'].astype('int64')

df['budget'] = df['budget'].astype('int64')

print(df.dtypes)
```

```python
# 1. Extract date and country safely
df[['date_part', 'country']] = df['released'].str.extract(r'(.+)\s\((.+)\)')

# 2. Convert with mixed format support
df['release_date'] = pd.to_datetime(df['date_part'], format='mixed', errors='coerce')

# 3. Extract year
df['release_year'] = df['release_date'].dt.year.astype('Int64')

# 4. Convert to dd-mm-yyyy (only where date exists)
df['release_date'] = df['release_date'].dt.strftime('%d-%m-%Y')

# 5. Drop temp column
df.drop(columns=['date_part'], inplace=True)

print(df[['released','release_date','release_year','country']].head())
```

```python
# sorting 
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```
